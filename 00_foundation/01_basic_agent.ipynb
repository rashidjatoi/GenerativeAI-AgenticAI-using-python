```python
# import dotenv pakcage to inject environment variables from .env file
from dotenv import load_dotenv
```

```python
# load environment variables from .env file
load_dotenv()
```

```python
# either we can import the key manually or it will be auto be managed by dotenv package
import os

# initilizing the value
open_api_keu = os.getenv("OPENAI_API_KEY")

# debugging purpose
print(open_api_keu)
```

```python
if open_api_keu is None:
    raise ValueError("OPENAI_API_KEY is not set in the environment variables.")
else : 
    print("OPENAI_API_KEY is set successfully.")
```

```python
# importing the openai api key
from openai import OpenAI

# creating an instance of the OpenAI class
client = OpenAI()
```

```python

# defining the question
question = "What is the capital of France?"

# defining the messages

messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": question}
    ]
```

```python
# creating a chat completion

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages
)
```

```python
# printing the response
print(response.choices[0].message.content)
```
